### 3. Data Corelation

```python
# Create correlation matrix 
sub.corr()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(15,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
# The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
plt.title('Correlation Heatmap')
plt.savefig(os.path.join(path1, 'Heatmap_Correlation'),bbox_inches='tight', pad_inches=0.1)
```

```python
# Create correlation heatmap using matplotlib
# Adding labels, a legend, and change the size of the heatmap

f = plt.figure(figsize = (18,18)) # figure size
plt.matshow(sub.corr(), fignum = f.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns, fontsize = 11, rotation = 47 ) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize = 11) # y axis labels
cb = plt. colorbar() # add a color legend called colorbar
cb.ax.tick_params(labelsize = 13) # add font size
plt.title('Correlation Matrix', fontsize = 17) # add title
plt.savefig(os.path.join(path1, 'Hatrix_Correlation'),bbox_inches='tight', pad_inches=0.1)
```

NOTE:
- Overall we can see low correlation on many variables.
- The ones that stand out are the rating values because they are all together considered in the overall rating. What IS interesting is that the location rating might has the lowest weight in the overall rating. Also isnt as strongly correlated to the other ratings.
- The "accomodation features" are also naturaly correlated. If I have more bedrooms and beds I also can accomodate more people and vise versa. Also it's higly likely that there are more bathrooms for three bedrooms than for a single bedroom.
- There is negative correlation between value rating and accomodates.
- There is a postive correlation between is superhost and overall rating
- strongest positive correlations: accomodates/beds,bedrooms,guest included and overall rating/value rating, accuracy rating, cleanliness rating
- strongest negative correlations: value rating/accomodates, beds and overall rating/accomodates, beds

### 8. Hypothesis:

```python
Alternative Hypothesis: If the price of an Airbnb listing is above 100 euros per night, it is more likely to have a good overall rating.
Null Hypothesis: There is no significant relationship between the price of an Airbnb listing above 100 euros per night and its overall rating.
Alternative Hypothesis: The greater the number of beds in an Airbnb listing, the higher the overall rating.
Null Hypothesis: There is no significant relationship between the number of beds in an Airbnb listing and the overall rating.
```
